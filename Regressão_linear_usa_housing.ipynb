```python
X = mainCasa[['avg_area_income', 'avg_area_house_age', 'avg_area_number_of_rooms',
       'avg_area_number_of_bedrooms', 'area_population']]

Y = mainCasa[['price']]
```

```python
from sklearn.model_selection import train_test_split as tts

X_train, X_test, Y_train, Y_test = tts(X, Y, train_size= 0.7, test_size = 0.3, random_state = 42)
```

```python
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
```

```python
lr.fit(X_train, Y_train)
```

```python
Y_pred = lr.predict(X_test)
```

```python
from sklearn.metrics import r2_score

r = r2_score(Y_test, Y_pred)
```

```python
print(f'r² = {r}')
```

```python
v = [i for i in range(1, 1500+1, 1)]

fig = plt.figure(figsize = (12, 8))
plt.plot(v, Y_test, color='blue')
plt.plot(v, Y_pred, color='red')
plt.xlabel('Index')
```
